## Part 3: Text organization

1. Import `pandas` and perform the tasks of data cleaning revised in the `Text Processing` notebook in order to get a Corpus and a Document-term Matrix, add or remove cleaning functions if necessary

```python
import pandas as pd
```

```python
data.keys()
```

```python
data_formatted = {key: [value] for (key, value) in data.items()}
pd.set_option('max_colwidth',1000)
data_df = pd.DataFrame.from_dict(data_formatted).transpose()
data_df.columns = ['report']
data_df = data_df.sort_index()
data_df.head()
```

```python
import re
import string
```

```python
def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('<.*?>', ' ', text)
    text = re.sub('\\n', ' ', text) 
    text = re.sub('\n', ' ', text) 
    text = re.sub('\t', ' ', text) 
    return text
```

```python
data_clean = pd.DataFrame(data_df.report.apply(clean_text))
```

```python
data_clean['company_name'] = [c[:-11] for c in data_clean.index]
```

**This is the Corpus**

```python
data_clean.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
data_cv = cv.fit_transform(data_clean.report)
data_tokens = cv.get_feature_names()
```

```python
data_matrix = pd.DataFrame(data_cv.toarray(), columns=data_tokens)
data_matrix.index = data_clean.index
```

**This is the Matrix**

```python
data_matrix.head()
```

## Part 4: Financial stop words

1. Write a process or function that will retrieve the top 30 used words per document and save it in the format of your preference

```python
import pandas as pd
```

```python
#data = pd.read_pickle('pickle/AnnualReports_matrix.pkl')
data = data_matrix.transpose()
data.head()
```

<br>
Find the top 30 words per report

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c]= list(zip(top.index, top.values))
```

```python
pd.DataFrame.from_dict(top_dict)
```

```python
# Let's first pull out the top 30 words for each company
words = []
for company in data.columns:
    top = [word for (word, count) in top_dict[company]]
    for t in top:
        words.append(t)
```

<br>
The main idea here is to filter or exclude such words from the financial jargon, wich are highly present on each document and presumibly do not represent any insigh, those could be added to a stop words list, like for instance "financial".
